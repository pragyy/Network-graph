## Making Raw data of desired format for visualization

```python
# import packages
import pandas as pd
```

```python
# reading dataset
df = pd.read_csv('as20000102.txt.gz')
df.head()
```

```python
# renaming the large and complex column name to easy name
df.columns = ['node']
```

```python
# splitting the info to "to node" and "from node" 
res = [] # storing the splitted value in this
for i in range(3, len(df)):
    val = df.loc[i, 'node']
    abc = val.split()
    res.append(abc)
```

```python
res
```

```python
# converting it to a new dataframe
new_df = pd.DataFrame(res, columns = ['node_1', 'node_2'])
new_df.head()
```

```python
# converting the df to csv file 
new_df.to_csv('abc.csv')
```
